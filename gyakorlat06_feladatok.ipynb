![](https://moodle.uni-corvinus.hu/theme/image.php/adaptable/atto_h5p/1709548862/icon)

````sql
SELECT  v.USERNEV,
        v.NEV,
        f.METTOL,
        f.MEDDIG,
        f.FELNOTT_SZAM + f.GYERMEK_SZAM as 'hány fő',
        SUM(f.FELNOTT_SZAM + f.GYERMEK_SZAM) OVER(PARTITION by f.ugyfel_fk
                                                  order by f.mettol
                                                  ROWS BETWEEN 3 preceding and 1 preceding) AS 'Előző 3 összesen'
FROM Foglalas f JOIN Vendeg v ON f.UGYFEL_FK=v.USERNEV
````

````sql
SELECT r.RAKTAR_NEV
FROM Raktar r JOIN Termek t ON t.RAKTAR_KOD=r.RAKTAR_KOD
WHERE t.KESZLET=(select MAX(t2.KESZLET)
                 from Termek t2)
GROUP BY r.RAKTAR_NEV
````

13. <span style="background-color: rgb(255, 255, 255); color: rgb(0, 0, 0); font-family: &quot;Open Sans&quot;, sans-serif; font-size: 14.44px;">Készítsünk listát (sorszámonként) az egyes rendelések teljes összegéről!&nbsp;</span> 

a. Szűrjünk azon tételekre, ahol ez az érték nagyobb, mint Viktor Éva rendeléseinek teljes összege! 

b. A teljes összeg oszlopot nevezzük el értelemszerűen!

````sql
SELECT rt.SORSZAM, SUM(rt.EGYSEGAR*rt.MENNYISEG) as 'Teljes összeg'
FROM Rendeles_tetel rt 
GROUP BY rt.SORSZAM
HAVING SUM(rt.EGYSEGAR*rt.MENNYISEG)>(select SUM(rt2.EGYSEGAR*rt2.MENNYISEG)
       from Rendeles_tetel rt2 JOIN Rendeles r2 ON r2.SORSZAM=rt2.SORSZAM
                    JOIN Ugyfel u2 ON u2.[LOGIN]=r2.[LOGIN]
        where u2.NEV='Viktor Éva')

````
